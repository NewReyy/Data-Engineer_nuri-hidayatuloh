### 1. Normalisasi Data: Normalisasi harga rumah dan harga per meter persegi menggunakan teknik Min-Max.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('./src/house_listings.csv')

# Mengubah format data pada kolom 'price'
df['price'] = df['price'].str.replace(' ', '').str.replace('.', '').astype(float)
df['price'] = df['price'].apply(lambda x: x / 1000 if x != 0 else x)

# Mengganti tanda spasi dan koma menjadi kosong dan mengubah tipe data kolom 'price_1m2' menjadi string
df['price_1m2'] = df['price_1m2'].str.replace(' ', '').str.replace(',', '').str.replace('.', '')

# Menghapus 'AZN/m²' dari kolom 'price_1m2' dan mengubah tipe data kolom tersebut menjadi float
df['price_1m2'] = df['price_1m2'].str.replace(r'(\D)', '', regex=True).astype(float)

# Membuat objek skaler dan melakukan normalisasi
scaler = MinMaxScaler()
df[['price', 'price_1m2']] = scaler.fit_transform(df[['price', 'price_1m2']])
```

```python
df.head()
```

### 2.Encoding Kategorikal: Terapkan encoding one-hot pada kolom 'Category', 'title_deed', 'repair', dan 'mortgage'.

```python
from sklearn.preprocessing import OneHotEncoder

# Membuat objek OneHotEncoder dengan parameter sparse=False
encoder = OneHotEncoder(sparse=False)

#Menghapus nilai NaN atau Null
df.dropna(subset=['category', 'title_deed', 'repair', 'mortgage'], inplace=True)

# Memanggil metode fit_transform pada objek OneHotEncoder dengan kolom yang akan di-encode
encoded_columns = encoder.fit_transform(df[['category', 'title_deed', 'repair', 'mortgage']])

# Mengubah hasil fit_transform menjadi DataFrame dengan menggunakan metode get_feature_names_out()
encoded_columns = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out())

# Menyatukan DataFrame hasil encoding dengan DataFrame asli menggunakan metode concat()
df = pd.concat([df, encoded_columns], axis=1)

# Menghapus kolom asli yang telah di-encode dari DataFrame
df.drop(['category', 'title_deed', 'repair', 'mortgage'], axis=1, inplace=True)
```

```python
df.head()
```

### 3. Aggregasi Data: Hitung rata-rata, median, dan mode harga rumah berdasarkan jumlah kamar dan tipe bangunan.

```python
df_grouped = df.groupby(['room_number', 'category_Köhnə tikili', 'category_Yeni tikili'])

# Hitung rata-rata, median, dan modus
df_aggregated = df_grouped['price'].agg(['mean', 'median', lambda x: x.mode().iloc[0]])

df_aggregated = df_aggregated.rename(columns={'<lambda_0>': 'modus'})
```

## 1. Mengelola Nilai yang Hilang pada kolom area dan floor

```python
# Menghitung jumlah nilai yang hilang pada kolom 'area'
num_missing_area = df['area'].isna().sum()

# Menghitung jumlah nilai yang hilang pada kolom 'floor'
num_missing_floor = df['floor'].isna().sum()

# Menampilkan jumlah nilai yang hilang
print(f"Jumlah nilai yang hilang pada kolom 'area': {num_missing_area}")
print(f"Jumlah nilai yang hilang pada kolom 'floor': {num_missing_floor}")
```

```python
# Memisahkan kolom 'floor' menjadi 'floor' dan 'total_floor'
df[['floor', 'total_floor']] = df['floor'].str.split(' / ', 1, expand=True)

# Menghapus karakter 'm²' pada kolom 'area'
df['area'] = df['area'].str.replace(' m²', '')

# Mengubah tipe data kolom 'floor', 'area', dan 'total_floor' menjadi numerik
df[['floor', 'area', 'total_floor']] = df[['floor', 'area', 'total_floor']].apply(pd.to_numeric, errors='coerce')

# Imputasi nilai yang hilang pada kolom 'area' dengan nilai rata-rata
rata_rata_area = df['area'].mean()
df['area'].fillna(rata_rata_area, inplace=True)

# Imputasi nilai yang hilang pada kolom 'floor' dengan nilai median
median_floor = df['floor'].median()
df['floor'].fillna(median_floor, inplace=True)
```

## 2. Mengatasi Outlier

```python
# Menampilkan informasi tambahan setelah menghapus outlier
print("Informasi sebelum menghapus outlier:")
df.head()
```

```python
# Menghitung kuartil pertama (Q1), kuartil ketiga (Q3), dan IQR (Interquartile Range)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Mengidentifikasi outlier
outlier_index = df[(df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR))].index

# Menghapus outlier
df_cleaned = df.drop(outlier_index)
```

## 3. Integritas Data

```python
# Verifikasi dan validasi kolom 'address'
# Pastikan semua nilai adalah string sebelum memeriksa karakter pertama
df['address_valid'] = df['address'].astype(str).apply(lambda x: x[0].isupper())

# Verifikasi dan validasi kolom 'url'
# Pastikan semua nilai adalah string sebelum memeriksa karakter pertama
df['url_valid'] = df['url'].astype(str).apply(lambda x: x.startswith('https://'))
```
