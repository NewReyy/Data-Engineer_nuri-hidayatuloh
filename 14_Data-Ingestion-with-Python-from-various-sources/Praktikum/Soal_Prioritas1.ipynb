## No. 3 Ambil data shipping menggunakan API using Library requests

```python
shipping_api_url = 'https://fakestoreapi.com/products'
shipping_response = requests.get(shipping_api_url)
shipping_data = shipping_response.json()

# Membuat DataFrame dari data pengiriman
shipping_df = pd.DataFrame(shipping_data)
```

```python
#Menampilkan dalam bentuk Dataframe
shipping_df.head(5)
```

### No. 4 Simpan semua data yang diingest ke dalam Pandas DataFrame dan ekspor ke file Parquet.

```python
# Menambahkan prefix untuk setiap kolom dari DataFrame yang diimpor
transactions_df = transactions_df.add_prefix('transactions_')
products_df = products_df.add_prefix('products_')
customers_df = customers_df.add_prefix('customers_')
shipping_df = shipping_df.add_prefix('shipping_')

merged_df = pd.concat([transactions_df, products_df, customers_df, shipping_df], axis=1)

# Menyimpan DataFrame ke file Parquet
merged_df.to_parquet('merged_data.parquet', index=False)
```
